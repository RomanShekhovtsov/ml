```python
# печать таблицы в HTML
def print_table( data):
    from IPython.display import HTML, display
    display(HTML(
        '<table><tr>{}</tr></table>'.format(
            '</tr><tr>'.join(
                '<td>{}</td>'.format('</td><td>'.join(str(_) for _ in row)) for row in data)
            )
     ))
```

```python
import re
import numpy as np

# Скачайте файл с предложениями (sentences.txt).
with open('sentences.txt') as f:
    sentences = list(f)

# Каждая строка в файле соответствует одному предложению. Считайте их, приведите каждую к нижнему регистру с помощью строковой функции lower().
sentences = map(lambda s: s.strip().lower(), sentences )

# Произведите токенизацию, то есть разбиение текстов на слова
sentences = map(lambda s: re.split('[^a-z]', s), sentences)

# Не забудьте удалить пустые слова после разделения.
sentences = list(map(lambda l: list(filter(lambda s: s!='', l)), sentences))

# Составьте список всех слов, встречающихся в предложениях.
# Сопоставьте каждому слову индекс от нуля до (d - 1), где d — число различных слов в предложениях.
d = dict()
i = 0
for s in sentences:
    for word in s:
        if d.get(word,'') == '':
            d[word] = i
            i += 1

# Создайте матрицу размера n * d, где n — число предложений. 
m = np.zeros((len(sentences), len(d)))

# Заполните ее: элемент с индексом (i, j) в этой матрице должен быть равен количеству вхождений j-го слова в i-е предложение.
for i in range(len(sentences)):
    sentence = sentences[i]
    for word in sentence:        
        j = d[word]
        m[i, j] = len(
            list(
                filter(lambda w: w==word, sentence)))

# У вас должна получиться матрица размера 22 * 254.
print(m.shape)
print(m[21])
#print(d)
```

```python
# Найдите косинусное расстояние от предложения в самой первой строке 
# (In comparison to dogs, cats have not undergone...) до всех остальных
# с помощью функции scipy.spatial.distance.cosine
from scipy.spatial import distance

distances = []
for i in range(len(sentences)):
    distances.append((
        ' '.join(sentences[i]), 
        distance.cosine(m[0, :], m[i, :]),
        i
    ))

# Какие номера у двух предложений, ближайших к нему по этому расстоянию 
# (строки нумеруются с нуля)?
distances.sort(key=lambda x: x[1])   
print_table(distances)
# 6 4

# Совпадают ли ближайшие два предложения по тематике с первым?
# - только у одного

# Совпадают ли тематики у следующих по близости предложений?
# - тоже только у одного
```
