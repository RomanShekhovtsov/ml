```python
# печать таблицы в HTML
def print_table( data, headers=None ):
    from IPython.display import HTML, display
    
    html = ''
    if not headers is None:
        html += '<tr><td><b>{}</b></td></tr>'.format('</b></td><td><b>'.join(str(_) for _ in headers))

    display(HTML(
        '<table><tr>{}</tr></table>'.format(
            html + '</tr><tr>'.join(
                '<td>{}</td>'.format('</td><td>'.join(str(_) for _ in row)) for row in data)
            )
     ))
    #print(html)
```

```python
from numpy.polynomial import polynomial

def approximate( x ):

    x = np.array( x )
    # Сформируйте систему линейных уравнений (то есть задайте матрицу коэффициентов A и свободный вектор b)
    # для многочлена первой степени, который должен совпадать с функцией f в точках 1 и 15.
    y = tuple( map(f, x) )
    print_table( zip(x, y), ('x', 'y') )

    A = []
    for i in range(len(x)):
        A.append( x ** float(i) )
    
    A = np.array(A).T
    print_table(A)
    b = y
 
    # Решите данную систему с помощью функции scipy.linalg.solve. 
    from scipy.linalg import solve
    w = solve( A, b )
    
    print("w: ", w)

    # Нарисуйте функцию f 
    xf = tuple( np.arange(0, 15.1, 0.1) )
    yf = tuple( map(f, xf) )
    plt.plot(xf, yf)

    # ... и полученный многочлен.
    polinom = lambda x: polynomial.polyval(x, w)
    y1 = list( map( polinom, xf ) )
    plt.plot( xf, y1 )

    plt.show()

approximate( (1, 15) )

# Хорошо ли он приближает исходную функцию?
# - так себе
```

```python
# Повторите те же шаги для многочлена второй степени, который совпадает с функцией f
# в точках 1, 8 и 15.
approximate( (1, 8, 15) )

# Улучшилось ли качество аппроксимации?
# - а то! :)
```
